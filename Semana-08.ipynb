## 3.- Sentencia Else (Sino)
Se encadena a un If para comprobar el caso contrario (en el que no se cumple la condición).

<center>
  <img width="450" src="https://raw.githubusercontent.com/jgomezm-tecsup/curso-introduccion-programacion/master/_assets/semana_08_if_else.png?token=AGVAKV2CSVCJNOZW737WZ7K7VKPWY" />
</center>


```python
n = 11
if n % 2 == 0:
    print(n,"es un número par")
else:
    print(n,"es un número impar")
```

## 4.- Sentencia Elif (Sino Si)
Se encadena a un if u otro elif para comprobar múltiples condiciones, siempre que las anteriores no se ejecuten.

<center>
  <img width="500" src="https://raw.githubusercontent.com/jgomezm-tecsup/curso-introduccion-programacion/master/_assets/semana_08_if_elif.png?token=AGVAKV767YLNR6KGFBCGAKK7VKQHC" />
</center>


```python
comando = "OTRA COSA"
if comando == "ENTRAR":
    print("Bienvenido al curso")
elif comando == "SALUDAR":
    print("Hola, espero que estes aprendiendo condcionales en Python")
elif comando == "SALIR":
    print("Saliendo del curso...")
else:
    print("Este comando no se reconoce")
```

```python
nota = float(input("Introduce una nota: "))
if nota >= 18:
    print("Sobresaliente")
elif nota >= 16:
    print("Notable")
elif nota >= 14:
    print("Bien")
elif nota >= 13:
    print("Suficiente")
else:
    print("Insuficiente")
```

Es posible simular el funcionamiento de elif con if utilizando expresiones condicionales

```python
nota = float(input("Introduce una nota: "))
if nota >= 18:
    print("Sobresaliente")
if nota >= 16 and nota < 18:
    print("Notable")
if nota >= 14 and nota < 16:
    print("Bien")
if nota >= 13 and nota < 14:
    print("Suficiente")
if nota < 13:
    print("Insuficiente")
```
