# 1.- Introducción
El nuevo tipo de "estructura de datos" que veremos son los "diccionarios".  Entre sus caracteristicas es que permite guardar datos que son asociados a una clave, lo cual nos permite encontrar rápidadmente los datos, como se entiende las llaves son similares a los índices de un libro y los valores serian los contenidos de los capítulos del libro.

Los Diccionarios se crean usando claves de la siguiente manera:

> a = **{** $ clave{1} : valor_{1}, clave_{2}: valor_{2}, clave_{3} : valor_{3},\dots $ **}**

Se puede notar que los "**:**" permiten separar las claves de los datos.


```
numeros1 = (1,2,3,4)   # Tupla --> No se puede cambiar los valores
numeros2 = [1,2,3,4]   # Lista --> Si se puede cambiar los valores
numeros3 = {1,2,3,4}   # Set.  --> No permite datos repetidos y se pueden cambiar

# Definición de un diicionario empleando números como clave
numeros4 = {1:10,2:20,3:30}  # Diccionario.
print(numeros4)
```

```
# Definición de un diicionario empleando cadenas como clave
estudiantes_c12 = {"nombre1":"Castro", "nombre2":"Jean Carlos","nombre3":"Renzo"}
print(estudiantes_c12)
```

# 2.- Propiedades 




### 2.1. Las "claves" en los diccionarios no se debe **repetir** porque generaria inconsistencias

```
# Sin claves repetidas
numeros = {1:10,2:20,3:30}
print(numeros)
# Con claves repetidas
numeros = {2:10,2:20,3:30}
print(numeros)
```

```
# Sin claves repetidas usando cadenas
colores = {"color-01":"rojo", "color-02":"verde", "color-03":"amarillo"}
print(colores)

# Con claves repetidas usando cadenas
colores = {"color-01":"rojo", "color-01":"verde", "color-03":"amarillo"}
print(colores)
```

### 2.2. Los diccionarios solo puede tener como claves valores booleanos, enteros, decimales y cadenas. No se permite otro tipo de datos


```
# Tiene como clave números booleanos
estados = {True:"SI",False:"NO"} 
print(estados)

# Tiene como clave números enteros
numeros = {1:10,2:20,3:20} 
print(numeros)

# Tiene como clave números decimales
decimales = {1.2:10,1.2:20,1.3:20} 
print(decimales)

# Tiene como clave cadenas
colores = {"rojo":1, "verde":2, "amarillo":3} 
print(colores)

# Tiene como claves tipos distintos
mixto = {True:"booleano",2:"entero", 4.5:"decimal","nombre":"cadena"}
print(mixto)
```

### 2.3.- En los diccionarios, los datos pueden ser accedido a traves de sus claves

```
# Tiene como clave números booleanos
estados = {True:"SI",False:"NO"} 
print(estados[True])  # accedo al daot con clave True
```

```
# Tiene como clave números enteros
numeros = {1:10,2:20,3:20} 
print(numeros[2])    # accedo al dato con clave 2
```

```
# Tiene como clave números decimales
decimales = {1.2:10,1.2:21,1.3:22} 
print(decimales[1.3])     # accedo al dato con clave 1,3
```

```
# Tiene como clave cadenas
colores = {"rojo":1, "verde":2, "amarillo":3} 
print(colores["verde"])     # accedo al dato con clave "verde"
```

```
# Tiene como claves tipos distintos
mixto = {True:"booleano",2:"entero", 4.5:"decimal","nombre":"cadena"}
print(mixto[4.5])       # accedo al dato con la clave 4.5
```

### 2.4.- En los diccionarios, se puede agregar o sobreescribir un registro a través de una clave, se usa la siguiente notación :
$$ diccionario[clave] = valor $$

```
mis_promedios  = {"matematica":14,
                  "lenguaje":16}
print(mis_promedios)

# Agrego el promedio de "programación"
mis_promedios["programacion"]= 17
print(mis_promedios)

# Modifico el promedio de "matematica"
mis_promedios["matematica"]= 13
print(mis_promedios)
```

### 2.5.- Los diccionarios permiten tener en sus datos tuplas, listas , set o diccionarios, es decir pueden tener un almacenamiento anidado

```
# Diccionario con listas anidadas
mis_notas   = {"matematica":[14,15,18,15],  
               "lenguaje":[20,18,15,14],
               "programacion":[13,11,10,15],}
print(mis_notas)

# Mostra la primera nota de Lenguaje
print(mis_notas["lenguaje"][0])
```

```
# Diccionario con diccionarios anidados
alumnos   = {"COD-01": {"nombre":"Jaime","apellido":"Gomez"},                          
             "COD-02": {"nombre":"Noemi","apellido":"Hernandez"},                                       
             "COD-03": {"nombre":"Julio","apellido":"Cesar"} }
print(alumnos)

# Mostra el apellido del alumno con codigo "COD-02"
print(alumnos["COD-02"]["apellido"])
```

# 3.- Funciones
Las Listas tiene funciones básicas que nos permiten realizar tareas concretas


## len()
Retorna el tamaño de un "set"

```
estudiantes = {1:"Juan",2: "Maria", 3:"ALberto",4:"Maribel",5: "Inmaculada"}
nro_estudiantes = len(estudiantes)
print(nro_estudiantes)
#print('El número  de estudiantes es = {}'.format(nro_estudiantes))
```

## update()

Permite agregar un diccionario a otro diccionario

```
# Diccionario unido a otro diccionario
colores = {1:"red", 3:"green", 5:"yellow", 7:"blue"}
nuevos_colores = {2:"white", 4:"black", 6:"purple"} # Es un set

#colores = colores + nuevos_colores # Esto da un error
colores.update(nuevos_colores)
print("colores :", colores)
```

## pop(clave)

Remueve un registro del diccionario por la clave y devuelve el valor del registro. En caso de no existir la clave da una excepción.

```
mis_notas = {"matematica":16,"lenguaje":15,"ingles":17}
print(mis_notas)

valor = mis_notas.pop("lenguaje") # remueve un registro por la clave
print("Valor leido {}".format(valor))

print(mis_notas)
```

## get(clave)

Lee el valor de un registro del diccionario por la clave y devuelve el valor del registro, no elimina el registro. En caso de no existir la clave devuelve como valor **None**

$$ diccionario[clave]$$

```
mis_notas = {"matematica":16,"lenguaje":15,"ingles":17}
print(mis_notas)

valor = mis_notas.get("lenguaje") # Lee un valor a través de una clave
print("Valor leido {}".format(valor))
print(mis_notas)

valor = mis_notas.get("programacion") # Lee un valor a través de una clave que no existe
print("Valor leido {}".format(valor))
print(mis_notas)

```

## in
Permite saber si una clave esta presente en un diccionario 

```
mis_notas = {"matematica":16,"lenguaje":15,"ingles":17}
print(mis_notas)

# ¿La nota de "programacion" esta en mis_notas?
esta_programacion = "programacion" in mis_notas
print(esta_programacion)

# ¿La nota de "lenguaje" esta en mis_notas?
esta_lenguaje = "lenguaje" in mis_notas
print(esta_lenguaje)
```

## sorted()
Permite ordenar las claves de un diccionario

```
mis_notas = {"matematica":16,"lenguaje":15,"ingles":17}
print(mis_notas)

# Ordena las claves de un diccionario
print(sorted(mis_notas))

# Ordena las claves de un diccionario en forma descendente
print(sorted(mis_notas, reverse=True))
```

## keys()
Obtiene las claves del diccionario

```
mis_notas = {"matematica":16,"lenguaje":15,"ingles":17}
print(mis_notas)

claves = mis_notas.keys()
print("Claves: {}".format(claves))
```

Otra forma de obtener las claves es a través de la función **list()**

```
mis_notas = {"matematica":16,"lenguaje":15,"ingles":17}
print(mis_notas)

claves = list(mis_notas)
print("Claves: {}".format(claves))
```

## items()
Obtiene los registros de un diccionario

```
mis_notas = {"matematica":16,"lenguaje":15,"ingles":17}
print(mis_notas)

registros = mis_notas.items()
print("Registros ", registros)
```

## values()
Obtiene los valores de un diccionario

```
mis_notas = {"matematica":16,"lenguaje":15,"ingles":17}
print(mis_notas)

valores = mis_notas.values()
print("Valores:{}".format(valores))
```

## clear()
Elimina todos los registros de un diccionario

```
mis_notas = {"matematica":16,"lenguaje":15,"ingles":17}
print(mis_notas)

mis_notas.clear() # remueve todos los registros
print(mis_notas)
```

## copy()
Hace una copia de un diccionario

```
mis_notas = {"matematica":16,"lenguaje":15,"ingles":17}
print("mis_notas =", mis_notas)

backup_mis_notas = mis_notas.copy() # hace una copia de respaldo
print("backup_datos =", backup_mis_notas)

print('Borra los registros de "mis_notas"')
mis_notas.clear()

print("mis_datos =", mis_notas)
print("backup_datos =", backup_mis_notas)
```

## type()

Permita saber el tipo de datos de una variable

```
mi_nro = 10
print(type(mi_nro)) 

mi_decimal = 10.4
print(type(mi_decimal)) 

es_martes = True
print(type(es_martes)) 

mi_nombre = "Jaime"
print(type(mi_nombre)) 

mi_tupla = ("JAIME","AAA","I","M","E",21,12.3,3)
print(type(mi_tupla)) 

mi_lista = ["JAIME","AAA","I","M","E",21,12.3,3]
print(type(mi_lista)) 

mi_set = {"JAIME","AAA","I","M","E",21,12.3,3}
print(type(mi_set)) 

mi_dicc = {"nombre":"Jaime",3:"martes", True:"Es positivo" }
print(type(mi_dicc)) 
```
