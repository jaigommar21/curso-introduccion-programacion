# **Laboratorio 08 : Estructuras de Control Condicional**


## 1.- Introducción

Las estructuras de control condicional permiten tomar decisiones. Se trata de estructuras muy importantes, ya que son las encargadas de controlar el flujo de un programa, según los requerimientos del programa.



## 2.- Sentencia If (Si)
Permite dividir el flujo de un programa en diferentes caminos. El if se ejecuta siempre que la expresión que comprueba devuelva True

```python
if True:  # equivale a if not False
    print("Se cumple la condición")
    print("También se muestre este print")
```

 Podemos encadenas diferentes If

```python
a = 5
if a == 2:
    print("a vale 2")
if a == 5:
    print("a vale 5")
```

O también anidar If dentro de If

```python
a = 5
b = 10
if a == 5:
    print("a vale",a)
    if b == 10:
        print("y b vale",b)
```

Como condición podemos evaluar múltiples expresiones, siempre que éstas devuelvan True o False

```python
if a==5 and b == 10:
    print("a vale 5 y b vale 10")
```

## 3.- Sentencia Else (Sino)
Se encadena a un If para comprobar el caso contrario (en el que no se cumple la condición).

```python
n = 11
if n % 2 == 0:
    print(n,"es un número par")
else:
    print(n,"es un número impar")
```

## 4.- Sentencia Elif (Sino Si)
Se encadena a un if u otro elif para comprobar múltiples condiciones, siempre que las anteriores no se ejecuten.

```python
comando = "OTRA COSA"
if comando == "ENTRAR":
    print("Bienvenido al curso")
elif comando == "SALUDAR":
    print("Hola, espero que estes aprendiendo condcionales en Python")
elif comando == "SALIR":
    print("Saliendo del curso...")
else:
    print("Este comando no se reconoce")
```

```python
nota = float(input("Introduce una nota: "))
if nota >= 18:
    print("Sobresaliente")
elif nota >= 16:
    print("Notable")
elif nota >= 14:
    print("Bien")
elif nota >= 13:
    print("Suficiente")
else:
    print("Insuficiente")
```

Es posible simular el funcionamiento de elif con if utilizando expresiones condicionales

```python
nota = float(input("Introduce una nota: "))
if nota >= 18:
    print("Sobresaliente")
if nota >= 16 and nota < 18:
    print("Notable")
if nota >= 14 and nota < 16:
    print("Bien")
if nota >= 13 and nota < 14:
    print("Suficiente")
if nota < 13:
    print("Insuficiente")
```

## 5.- Instrucción Pass
Sirve para finalizar un bloque, se puede utilizar en un bloque vacío.

```python
if True:
    pass
```

## 6.- Conclusiones
Las estructuras de control condicional nos permite controlar los flujos de un programa de acuerdo a los requerimientos que deben cumplir el programa.
